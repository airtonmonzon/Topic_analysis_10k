## Analisis de Topics

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel('corpus_results.xlsx', index_col=0)

df['year'] = df['year'] - 1
df.shape
```

```python
# Define el numero de topico y el nombre 
topic_num = '09'  # Ejemplo: cambia a '02', '03', etc.para cada topico
topic_name = 'Hoteles y Casinos'  # Elige el nombre del topico
```

```python
topic_column = f'topic_{topic_num}'

sns.set_style("whitegrid")
palette = sns.color_palette("husl", n_colors=len(df['sector'].unique()))

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='year', y=topic_column, hue='sector', marker='o', palette=palette, linewidth=2.5, markersize=8, ci=None)

plt.title(f'{topic_name}', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Topic Value (%)', fontsize=12)
plt.legend(title='Sector', title_fontsize=12, fontsize=10, frameon=True, facecolor='white', edgecolor='gray')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(ticks=df['year'].unique(), fontsize=10, rotation=0)

plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))

plt.tight_layout()
plt.savefig(f'{topic_name}_Trends.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
selected_sector = 'Communication Services'  # Change this to the desired sector

df_sector = df[df['sector'] == selected_sector]

topic_names = {
    'topic_01': 'Consumo y Logstica',
    'topic_02': 'Marketing Digital',
    'topic_03': 'Informacin Tecnologica',
    'topic_04': 'Media ',
    'topic_05': 'Finanzas',
    'topic_06': 'Semiconductores y desarrollo de hardware',
    'topic_07': 'Automotriz',
    'topic_08': 'Telecomunicaciones y conectividad',
    'topic_09': 'Hoteles y Casinos'
}  

sns.set_style("whitegrid")  # Clean grid background
palette = sns.color_palette("husl", n_colors=9)  # 9 colors for 9 topics

plt.figure(figsize=(12, 6))  # Larger figure for better readability

for i in range(1, 10):
    topic_column = f'topic_{i:02d}'  # Formats as 'topic_01', 'topic_02', etc.
    sns.lineplot(data=df_sector, x='year', y=topic_column, marker='o', linewidth=2.5, markersize=8, ci=None, label=topic_names[topic_column])

plt.title(f'Tendencia de topicos por Sector: {selected_sector}', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Topic Value', fontsize=12)
#plt.legend(title='Topics', title_fontsize=12, fontsize=10, frameon=True, facecolor='white', edgecolor='gray')
plt.legend(title='Topics', title_fontsize=12, fontsize=10, frameon=True, facecolor='white', edgecolor='gray', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)

plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))

plt.xticks(ticks=df_sector['year'].unique(), fontsize=10, rotation=0)

plt.tight_layout()
plt.savefig(f'Topic_Trends_{selected_sector}.png', dpi=300, bbox_inches='tight')  # Save as high-resolution image
plt.show()
```

```python
selected_companies = ['Nvidia','Alphabet_Inc','Meta_Platforms','Tesla_Inc','Amazon','Microsoft','Apple_Inc']  # Cambia esta lista a las empresas deseadas

df_selected = df[df['company'].isin(selected_companies)]

topic_names = {
    'topic_01': 'Consumo y Logstica',
    'topic_02': 'Marketing Digital',
    'topic_03': 'Informacin Tecnologica',
    'topic_04': 'Media',
    'topic_05': 'Finanzas',
    'topic_06': 'Semiconductores y desarrollo de hardware',
    'topic_07': 'Automotriz',
    'topic_08': 'Telecomunicaciones y conectividad',
    'topic_09': 'Hoteles y Casinos'
}  

sns.set_style("whitegrid")
palette = sns.color_palette("husl", n_colors=9)

plt.figure(figsize=(12, 6))

for i in range(1, 10):
    topic_column = f'topic_{i:02d}'
    # Calcular promedio por ao para las empresas seleccionadas
    df_topic = df_selected.groupby('year')[topic_column].mean().reset_index()
    sns.lineplot(data=df_topic, x='year', y=topic_column, marker='o', linewidth=2.5, markersize=8, ci=None, label=topic_names[topic_column])

plt.title(f'Tendencia de tpicos de las Magnificent 7', fontsize=14, pad=15, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Topic Value', fontsize=12)
plt.legend(title='Topics', title_fontsize=12, fontsize=10, frameon=True, facecolor='white', edgecolor='gray', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))

plt.xticks(ticks=df_selected['year'].unique(), fontsize=10, rotation=0)

plt.tight_layout()
plt.savefig(f'Topic_Trends_{"_".join(selected_companies)}.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick  # Para formatear el eje Y como %

topic_names = {
    'topic_01': 'Consumo y Logstica',
    'topic_02': 'Marketing Digital',
    'topic_03': 'Informacin Tecnolgica',
    'topic_04': 'Media',
    'topic_05': 'Finanzas',
    'topic_06': 'Semiconductores y desarrollo de hardware',
    'topic_07': 'Automotriz',
    'topic_08': 'Telecomunicaciones y conectividad',
    'topic_09': 'Hoteles y Casinos'
}

custom_palette = {
    'Consumo y Logstica': '#1f77b4',
    'Marketing Digital': '#ff7f0e',
    'Informacin Tecnolgica': '#2ca02c',
    'Media': '#d62728',
    'Finanzas': '#9467bd',
    'Semiconductores y desarrollo de hardware': '#8c564b',
    'Automotriz': '#e377c2',
    'Telecomunicaciones y conectividad': '#7f7f7f',
    'Hoteles y Casinos': '#bcbd22'
}

topic_cols = [f'topic_{i:02d}' for i in range(1, 10)]

df_company_topic_avg = df_selected.groupby('company')[topic_cols].mean()

df_company_topic_avg['max_topic'] = df_company_topic_avg[topic_cols].idxmax(axis=1)
df_company_topic_avg['max_topic_score'] = df_company_topic_avg.apply(lambda row: row[row['max_topic']], axis=1)
df_company_topic_avg['max_topic_name'] = df_company_topic_avg['max_topic'].map(topic_names)

df_plot = df_company_topic_avg.reset_index()[['company', 'max_topic_name', 'max_topic_score']]

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=df_plot, x='company', y='max_topic_score', hue='max_topic_name',
                 dodge=False, palette=custom_palette)

for container in ax.containers:
    ax.bar_label(container, labels=[f'{h:.1%}' for h in container.datavalues],
                 label_type='edge', padding=3, fontsize=10)

ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

plt.title("Tpico ms dominante por empresa", fontsize=14)
plt.ylabel("Promedio del Tpico dominante (%)", fontsize=12)
plt.xlabel("Empresa", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Tpico", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd

magnificent_7 = ['Nvidia','Microsoft','Apple_Inc','Tesla_Inc','Alphabet_Inc','Meta_Platforms','Amazon']
df_m7 = df[df['company'].isin(magnificent_7)].copy()

topic_names = {
    'topic_01': 'Consumo y Logstica',
    'topic_02': 'Marketing Digital',
    'topic_03': 'Informacin Tecnolgica',
    'topic_04': 'Media',
    'topic_05': 'Finanzas',
    'topic_06': 'Semiconductores y desarrollo de hardware',
    'topic_07': 'Automotriz',
    'topic_08': 'Telecomunicaciones y conectividad',
    'topic_09': 'Hoteles y Casinos'
}

all_years = sorted(df_m7['year'].unique())
all_combinations = pd.MultiIndex.from_product([magnificent_7, all_years], names=['company', 'year'])
df_filled = df_m7.set_index(['company', 'year']).reindex(all_combinations).reset_index()

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
axes = axes.flatten()

for idx, company in enumerate(magnificent_7):
    df_company = df_filled[df_filled['company'] == company]
    ax = axes[idx]
    
    for i in range(1, 10):
        topic_col = f'topic_{i:02d}'
        if topic_col in df_company:
            ax.plot(df_company['year'], df_company[topic_col], label=topic_names[topic_col], marker='o', linewidth=1.5)

    ax.set_title(company, fontsize=11)
    ax.set_xlabel("Ao")
    ax.set_ylabel("Valor del tpico (%)")
    ax.set_xticks(all_years)
    ax.set_xticklabels(all_years, rotation=45)
    ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))  # Eje Y en porcentaje
    ax.grid(True)

if len(magnificent_7) < len(axes):
    fig.delaxes(axes[-1])

handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', bbox_to_anchor=(0.5, -0.1), ncol=3, title='Tpicos')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig("Magnificent7_Topic_Trends_All.png", dpi=300)
plt.show()

```

## Indice Bursatil de las Magnifient 7

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

magnificent_7 = ["AAPL", "MSFT", "GOOGL", "AMZN", "NVDA", "META", "TSLA"]
benchmark = ["SPY"]
tickers = magnificent_7 + benchmark

data = yf.download(tickers, start="2015-01-01", group_by='ticker', auto_adjust=True)

adj_close = pd.DataFrame({ticker: data[ticker]['Close'] for ticker in tickers})

normalized = adj_close / adj_close.iloc[0]

normalized["Magnificent 7 (Promedio)"] = normalized[magnificent_7].mean(axis=1)

plot_data = normalized[["Magnificent 7 (Promedio)", "SPY"]].rename(columns={"SPY": "S&P 500"})

plot_df = plot_data.reset_index().melt(id_vars="Date", var_name="Compaa", value_name="Retorno acumulado")
plt.figure(figsize=(14, 7))
sns.lineplot(data=plot_df, x="Date", y="Retorno acumulado", hue="Compaa", linewidth=2.2)

plt.title("Retorno Acumulado: Magnificent 7 (Promedio) vs S&P 500 (desde 2015)", fontsize=16)
plt.xlabel("Fecha")
plt.ylabel("Retorno acumulado")
plt.legend(title="Serie")
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

sns.set(style="whitegrid")
sector_etfs = {
    "Informacin Tecnolgica": "XLK",
    "Salud": "XLV",
    "Financiero": "XLF",
    "Consumo Discrecional": "XLY",
    "Industriales": "XLI",
    "Consumo Bsico": "XLP",
    "Energa": "XLE",
    "Materiales": "XLB",
    "Utilities": "XLU",
    "Inmobiliario": "XLRE",
    "Servicios de Comunicacin": "XLC"
}

# Precios ajustados
tickers = list(sector_etfs.values())
data = yf.download(tickers, start="2020-01-01", auto_adjust=True)["Close"]

start_date = data.index[0]
end_date = data.index[-117]
years = (end_date - start_date).days / 365

cagr = ((data.iloc[-1] / data.iloc[0]) ** (1 / years) - 1)
cagr = cagr.rename({v: k for k, v in sector_etfs.items()}).sort_values(ascending=False)
cagr = cagr.reset_index()
cagr.columns = ["Sector", "CAGR (%)"]
cagr["CAGR (%)"] = (cagr["CAGR (%)"] * 100).round(2)

plt.figure(figsize=(12, 7))
sns.barplot(data=cagr, x="CAGR (%)", y="Sector", palette="crest")

plt.title(f"Tasa de Crecimiento Anual Compuesta (CAGR) por Sector del S&P 500\nDesde {start_date.date()} hasta {end_date.date()}", fontsize=16)
plt.xlabel("CAGR (%)")
plt.ylabel("Sector")
plt.tight_layout()
plt.show()
```
